# Datenvisualisierung und Erkundung

In diesem Notebook werden wir eine explorative Datenanalyse (EDA) durchführen, um die Daten zu verstehen und Zusammenhänge zwischen verschiedenen Merkmalen zu erkennen. Der erste Schritt besteht darin, eine Verbindung zur Apache Cassandra Datenbank herzustellen und auf unseren spezifischen Keyspace zuzugreifen, der unsere Tabellen enthält. Diese Verbindung wird durch den Cassandra-Cluster und die Session, die wir erstellen, ermöglicht.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from cassandra.cluster import Cluster
import seaborn as sns
```

```python
# Connect to Cassandra
cluster = Cluster(['cassandra'])
session = cluster.connect('oulad') 
```

Nachdem wir uns mit der Datenbank verbunden haben, können wir beginnen, die Daten zu laden und zu analysieren. Dazu nutzen wir verschiedene Arten von Diagrammen, darunter Histogramme, Balkendiagramme und Boxplots. Histogramme und Balkendiagramme sind nützlich, um die Verteilung von Merkmalen zu verstehen, während Boxplots uns helfen können, Ausreißer zu identifizieren.

Darüber hinaus untersuchen wir auch spezifische Untermengen der Daten. Zum Beispiel analysieren wir die Daten von Studierenden, die ihr Studium abgebrochen haben, und berechnen den Prozentsatz der Studierenden, die abgebrochen haben. Dies hilft uns, ein besseres Verständnis für die Zielvariable zu bekommen.

Letztendlich zielt die EDA darauf ab, uns dabei zu helfen, die zugrunde liegenden Muster in den Daten zu entdecken und Hypothesen für die weitere Analyse zu entwickeln.

```python
# Load student_info from Cassandra
query = "SELECT * FROM student_info"
rows = session.execute(query)
student_info = pd.DataFrame(list(rows))
```

```python
# Basic information about the dataset
print(student_info.info())
print(student_info.describe())
```

```python
# Countplot of genders with labels
plt.figure(figsize=(6, 4))
student_info['gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for number of previous attempts
plt.figure(figsize=(6, 4))
student_info['num_of_prev_attempts'].plot(kind='hist', edgecolor='black')
plt.title('Number of Previous Attempts Distribution')
plt.xlabel('Number of Previous Attempts')
plt.show()
```

```python
# Boxplot for number of previous attempts
plt.figure(figsize=(6, 4))
student_info.boxplot(column=['num_of_prev_attempts'])
plt.title('Number of Previous Attempts Distribution')
plt.show()
```

```python
# Analysis of withdrawn students
withdrawn_students = student_info[student_info['has_withdrawn'] == True]
print(withdrawn_students.describe())
```

```python
# Percentage of students who have withdrawn
percentage_withdrawn = len(withdrawn_students) / len(student_info) * 100
print(f'Percentage of students who have withdrawn: {percentage_withdrawn:.2f}%')
```

```python
# Load the data from the student_vle table
query = "SELECT * FROM student_vle"
rows = session.execute(query)
student_vle = pd.DataFrame(list(rows))
```

```python
# Explore the student_vle data
print(student_vle.info())
print(student_vle.describe())
```

```python
# Histogram for sum_click in the student_vle table
plt.figure(figsize=(6, 4))
student_vle['sum_click'].plot(kind='hist', edgecolor='black')
plt.title('Distribution of Sum of Clicks in VLE')
plt.xlabel('Sum of Clicks')
plt.show()
```

```python
query = "SELECT * FROM student_vle_extended"
rows = session.execute(query)
```

```python
student_vle_extended = pd.DataFrame(list(rows))
```

```python
# Explore the student_vle_withdrawn data
print(student_vle_extended.info())
print(student_vle_extended.describe())
```

```python
sns.countplot(x='code_module', hue='has_withdrawn', data=student_vle_extended)
plt.title('Abbruchraten nach Modul')
plt.show()
```

```python
sns.countplot(x='code_presentation', hue='has_withdrawn', data=student_vle_extended)
plt.title('Abbruchraten nach Präsentation')
plt.show()
```

```python
sns.boxplot(x='has_withdrawn', y='sum_click', data=student_vle_extended)
plt.title('Beziehung zwischen Klickanzahl und Studienabbruch')
plt.show()
```

Am Ende der explorativen Datenanalyse trennen wir die Verbindung zur Datenbank, um Ressourcen freizugeben

```python
session.shutdown()
cluster.shutdown()
```
